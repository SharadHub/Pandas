Visualizing Data

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (simulate a mini avocado dataset)
data0 = {
    'Date': [
        '2023-01-01', '2023-01-08', '2023-01-15',
        '2023-02-01', '2023-02-08', '2023-2-15',
        '2023-03-01', '2023-03-08', '2023-03-15'
        ],
    'Small Hass': [
        15000, 18000, 16000, 14000, 19000, 17000, 21000, 22000, 23000
        ],
    'Large Hass': [
        5000, 6000, 7000, 8000, 9000, 10000, 5000, 6000, 7000
        ],
    'Extra Large Hass': [
        3000, 3500, 4000, 6000, 7500, 8000, 3000, 3500, 4000
        ]
}
df0 = pd.DataFrame(data0)
df0['Date'] = pd.to_datetime(df0['Date'])
print(df0)

df0.set_index('Date').plot(kind='line', marker='o', colormap='viridis')
plt.title("Avocado Sales Trends by Size Over Time")
plt.xlabel("Date")
plt.ylabel("Sales Volume")
plt.grid(True)
plt.show()
```

Using Matplotlib with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.DataFrame({
    'size': ['small', 'medium', 'large'],
    'count': [100, 150, 75]
})

df1.plot(kind='bar', x='size', y='count')
plt.show()
```

1. Which avocado size is most popular?

```python
data1 = {
    'size': ['small', 'medium', 'large', 'medium', 'small', 'large', 'large'],
    'total_sold': [100, 150, 200, 160, 90, 210, 190]
}
avocados = pd.DataFrame(data1)
grouped = avocados.groupby('size')['total_sold'].sum()


grouped.sort_values(ascending=False).plot(kind='bar', color='blue')
plt.title("Popularity of avocados sizes")
plt.xlabel('Avocados size')
plt.ylabel('Total Sold')
plt.grid(axis='y')
plt.show()
```

Changes in sales over Time

```python
data2 = {
    'date' : ['2021-01-01', '2021-01-01', '2021-01-02', '2021-01-02'],
    'total_sold' : [200, 300, 250, 280]
}
df2 = pd.DataFrame(data2)

df2['date'] = pd.to_datetime(df2['date'])

daily_sales = df2.groupby('date')['total_sold'].sum()

daily_sales.plot(kind='line', marker='o', color='green')
plt.title('Daily Avocado Sales')
plt.xlabel('Date')
plt.ylabel('Total Sold')
plt.xticks(rotation=45)
plt.show()
```

```python
data3 = {
    'avg_price' : [1.0, 1.2, 0.8, 1.5, 1.1],
    'total_sold1' : [100, 800, 1200, 600, 950]
}

df3 = pd.DataFrame(data3)

plt.scatter(df3['avg_price'], df3['total_sold1'], color = 'orange')
plt.title('Supply and Demand : Price v/s Sold')
plt.xlabel('Average Price')
plt.ylabel('Total Sold')
plt.show()
```

Price of Conventional v/s Organic Avocados

```python
avocados = pd.DataFrame(
    {
        'type' : ['conventional', 'organic', 'conventional', 'organic'],
        'avg_price' : [1.0, 1.5, 1.1, 1.6]
    }
)
avg_price_by_type = avocados.groupby('type')['avg_price'].mean()

avg_price_by_type.plot(kind='bar', color=['blue', 'green'])
plt.title('Average Price : Organic v/s Conventional')
plt.ylabel('Average Price')
plt.show()
```
